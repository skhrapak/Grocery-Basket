# 0.4 Looking for loyal customers

```python
#Creating max_order column
df_ords_prods_merge['max_order']= df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(15)
```

## Now use loc() function to create a flag

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order']>40, 'loyality_flag'] ='Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order']> 10), 'loyality_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order']<=10, 'loyality_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyality_flag'].value_counts(dropna = False)
```

# 0.3 Spending habits of customer categories

```python
df_ords_prods_merge.groupby('loyality_flag').agg({'prices':['sum']})
```

## Regular customers nuy more than other categories, which probably makes sense. Loyal customer bring more than new customers. 

# 0.4. Spending flag for customers

```python
#Creating new column - avg_order
df_ords_prods_merge['avg_order']= df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.average)
```

```python
df_ords_prods_merge[['user_id','avg_order']].head(5)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_order']>=10, 'spending_flag'] ='High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_order']<10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## There are much more low spender, which is reasonable 

```python
df_ords_prods_merge[['user_id','avg_order', 'spending_flag']].head(1000)
```

# 0.4 Frequecy order flag for customers

```python
#Creating new column - median_days_from_previous_order
df_ords_prods_merge['median_days_from_previous_order']= df_ords_prods_merge.groupby(['user_id'])['days_from_previous_order'].transform(np.median)
```

```python
df_ords_prods_merge[['user_id','median_days_from_previous_order']].head(100)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_from_previous_order']>20, 'frequency_flag'] ='Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_from_previous_order'] <= 20) & (df_ords_prods_merge['max_order']> 10), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_from_previous_order']<=10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge['frequency_flag'].value_counts(dropna = True)
```
